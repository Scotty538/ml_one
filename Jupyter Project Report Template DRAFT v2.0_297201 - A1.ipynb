### Task 3.a 

Which tournaments have had on average the most upsets (where a lower-ranked player defeated a higher- ranked player)? List the top 10 and plot their averages.

I need to count the total number of upsets for each tournament, how many times each tournament has been held over the period, and then use these values to calculate the average number of upsets per tournament.

We will need a binary indicator to decide if a match was an upset or not:

```python
def is_upset(rank_diff):
    if rank_diff > 0:
        return 1
    else:
        return 0

tennis_total_clean['Upset'] = tennis_total_clean['Rank_diff'].apply(is_upset)
# Didn't need axis as Rank-diff is a vector
```

Now let's create a streak counter function:

```python
def streak_counter(winner):
    count = 1
    max_count = 0
    for index in range(1,len(winner)):
        if winner.loc[index -1] == winner.loc[index]:
            count += 1
        else:
            if count > max_count:
                max_count = count
            count = 1
    return max_count
```
