### Task 1.d 

What is the most common tournament surface? If a tennis player wanted to be competitive at the majority of tournaments, what type of surface should they practice most on?

```python
surface = tennis_total_clean[['Surface', 'Tournament']].groupby('Surface').nunique()
surface = surface.reset_index()
surface = surface.sort_values('Tournament')
surface
```

```python
surface.plot(kind = 'bar', x = 'Surface', y = 'Tournament', legend = False)
plt.xlabel("Surface Type")
plt.ylabel("Count")
plt.title("Types of Tournament Surfaces")
plt.show()
```

```python
An aspiring tennis player should spend most of their time practicing on hard courts.
```

```python
Which tournament has had the most retirements? 
```

```python
retirements = tennis_total_clean[['Comment', 'Tournament']].groupby(['Tournament', 'Comment']).value_counts()
retirements = retirements.reset_index()
retirements = retirements.rename(columns={"count": "num_retirements"})
retirements = retirements[['Tournament','num_retirements']][retirements['Comment']=='Retired'].sort_values(by='num_retirements', ascending = False).head(5)
retirements
```

```python
retirements.plot(kind = 'bar', x = 'Tournament', y = 'num_retirements', rot = 45, legend = False)
plt.xlabel("Tournaments")
plt.ylabel("Number of Retirements")
plt.title("Most Dangerous Tournaments")
plt.show()
```

With 75, the US Open has the most retirements, with almost double the number of the next most dangerous tournament which is the Australian Open (38).
